```python
#using label encoder
categorical_columns = ['HomePlanet', 'CryoSleep', 'Destination', 'Transported']

categorical_columnss = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP']

label_encoder = LabelEncoder()
#train
for col in categorical_columns:
    train[col] = label_encoder.fit_transform(train[col])
#test
for col in categorical_columnss:
    test[col] = label_encoder.fit_transform(test[col])
```

```python
from sklearn.preprocessing import StandardScaler

X = train.drop('Transported', axis = 1)#feature data set 
Y= train['Transported']

#scaler
X = StandardScaler().fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.2)
clf = RandomForestClassifier( random_state=44) 

#train the model
clf.fit(X_train, Y_train)
```

```python
predictions =clf.predict(X_test)
from sklearn.metrics import classification_report

print( classification_report(Y_test, predictions) )
```
